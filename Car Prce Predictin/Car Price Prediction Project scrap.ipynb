```python
import json
import requests
import pandas as pd
import time
```

```python
URL = 'https://www.olx.in/api/relevance/v2/search'
```

```python
df = pd.DataFrame()
id_list = []
```

```python
def fetch_data_from_olx(page):
    
    all_list = []
    
    PARAMS = {
        'category': '5',
        'facet_limit': '100',
        'lang': 'en-IN',
        'location': '1000001',
        'location_facet_limit': '20',
        'page': page + 1,
        'platform': 'web-desktop',
        'user': '17e058d8ca7x378152c',
    }

    headers = {
        'User-Agent': 'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_10_1) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/39.0.2171.95 Safari/537.36'
    }




    # sending get request and saving the response as response object
    response = requests.get(url = URL,headers=headers,params=PARAMS)

    # extracting data in json format
    response_json = response.json()
        


    try:
        for one_data in response_json['data']:

            temp_dict = {}

            temp_dict['id'] = one_data['id']

            temp_dict['url'] = one_data['title']

            temp_dict['location'] = one_data['locations_resolved']['ADMIN_LEVEL_1_name']
            temp_dict['car_body_type'] = one_data['car_body_type']
            temp_dict['price'] = one_data['price']['value']['raw']


            for i in range(len(one_data['parameters'])):
                key = 'fuel' if one_data['parameters'][i]['key'] == 'petrol' else one_data['parameters'][i]['key']
                key = 'number_of_owners' if one_data['parameters'][i]['key'] == 'first_owner' else one_data['parameters'][i]['key']
                temp_dict[key] = one_data['parameters'][i]['value']

            if temp_dict['id'] not in id_list:
                all_list.append(temp_dict)
                id_list.append(temp_dict['id'])
                continue
            
    except Exception as e:
        print(response_json)
        print(e)

    return all_list
```

```python
i = 0
while True: 
    try:
        all_list = fetch_data_from_olx(i)
        df = df.append(pd.DataFrame(all_list))
        i+=1
        
    except Exception as e:
        print(i,df.shape)
        print(e)
        time.sleep(10)

    print(df.shape[0],i)
    
    if i >= 25:
        i = 0
    
    if df.shape[0] > 6000:
        break
df.shape
```

```python
df.to_csv('5000_old_car.csv')
```
